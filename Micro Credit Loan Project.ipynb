```python
# Removing 2 column form data

df_new = df.drop(columns=['pcircle','pdate'])
df_new.shape          # new shape
```

```python
df_new.plot(kind='box',subplots=True,figsize=(15,35), layout=(8,5),sharex= False, sharey= False)
plt.show()
```

```python
df_new.hist(figsize=(15,35),layout=(8,5))
plt.show()
```

 So from boxplot and histogaram also, we found that there are outliers in our data

```python
# Now removing outliers from data by z-score method.

from scipy import stats

df_z=df_new[(np.abs(stats.zscore(df_new.drop(columns=['msisdn']))) < 3).all(axis=1)]     # droping column 'msisdn' ust for calculating z-score, as it is of string type
print(df_new.shape)          # old data shape
print(df_z.shape)            # new data shape 
```

Our data is now free from outliers but we've lost almost 40% of data.

```python
df_z.hist(figsize=(15,35),layout=(8,5))
plt.show()
```

```python
#countplot
sns.countplot(df_z['label'])            
```

## Another method for treating outliers
as z-score method delets around 40% of our data

```python
# Now removing outliers from data by MEDIAN method.

df_med=df_new.drop(columns=['msisdn','label'])           # as msisdn does not have any impact on our model and 'label' columns has just 2 variable which will add later.
df_med.shape
```

```python
for i in df_med.columns:
    df_med[i] = np.where(df_med[i] > (df_med[i].quantile(0.95)), (df_med[i].quantile(0.50)), df_med[i])
```

```python
df_med.hist(figsize=(15,35),layout=(8,5))         # Visualizing outliers
plt.show()
```

```python
df_median=pd.concat([df_med,df['label']],axis=1)           # for joining label column to the median treated data to make a whole dataframe.

print(df_med.shape)                 
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df_median.corr(),annot=True)
plt.show()
```

As we see that our data is highly IMBALANCED, so we have to balance that first.

```python
# but before that, lets divide the data into input and output variables,

x = df_median.drop(columns=['label'])
y = df_median['label']
print(x.shape)
print(y.shape)
```

```python
# now installing imblearn for treating imbalance dataset.

!pip install imblearn
```

```python
# Balancing data by SMOTE technique.

from imblearn.combine import SMOTETomek
```

```python
smt = SMOTETomek(random_state=45)
x_res,y_res = smt.fit_sample(x,y)

print(x_res.shape)
print(y_res.shape)
```

```python
# Lets check now is or data balanced or not.

sns.countplot(y_res) 
```

So now from above countplot we can  see that our data is balaced now and can proceed further.

```python
# Now lets normalise the data

from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
x_res = ss.fit_transform(x_res)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,auc

x_res_train,x_res_test,y_res_train,y_res_test=train_test_split(x_res,y_res,test_size=0.25,random_state=42)
```

```python
print(x_res_train.shape,x_res_test.shape)
print(y_res_train.shape,y_res_test.shape)
```

# Hyperparameter tuning

```python
# To improve our best models till now, we use Grid search cv to find out which parameters are best for these models.

from sklearn.model_selection import GridSearchCV

# parameters for DTC
params_dtc = {'criterion':('gini', 'entropy'), 'max_depth':(4,6,8,12)}   

# parameters for RFC,GBC
params_bb = {'n_estimators':[100,500],'random_state':[10,100]}

# for Decision tree classifier

dtc = GridSearchCV(DT,param_grid=params_dtc,cv=10,scoring='accuracy')
dtc.fit(x_res_train,y_res_train)
print("Best parameters for DecisionTreeClassifier",dtc.best_params_) 
```

```python
# For Random forest classifier

rfc = GridSearchCV(RFC,param_grid=params_bb,scoring='accuracy')
rfc.fit(x_res_train,y_res_train)
print("Best parameters for RandomForestClassifier: ",rfc.best_params_)
```

```python
# for Gradient boosting classifier

gbc = GridSearchCV(GBC,param_grid=params_bb,scoring='accuracy')
gbc.fit(x_res_train,y_res_train)
print("Best parameters for GradientBoostingClassifier: ",gbc.best_params_)
```

# Now applying parameters found from Grid Search and again finding these algorithms.
